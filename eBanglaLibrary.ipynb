```
#@title Do not Edit. Epub Generator
#@markdown wait untill complete
!pip install ebooklib
import os

import requests
from shutil import copyfile
from bs4 import BeautifulSoup

from ebooklib import epub
import random

import os.path
from os import path

import io

from google.colab import output
output.clear()

#@title Parser
t =[]
fn= []
c = []

chapter_title_number = ['০০', '০১', '০২', '০৩', '০৪', '০৫', '০৬', '০৭', '০৮', '০৯', '১০', '১১', '১২', '১৩', '১৪', '১৫',
                        '১৬', '১৭', '১৮', '১৯', '২০', '২১', '২২', '২৩', '২৪', '২৫', '২৬', '২৭', '২৮', '২৯', '৩০', '৩১',
                        '৩২', '৩৩', '৩৪', '৩৫', '৩৬', '৩৭', '৩৮', '৩৯', '৪০', '৪১', '৪২', '৪৩', '৪৪', '৪৫', '৪৬', '৪৭',
                        '৪৮', '৪৯', '৫০', '৫১', '৫২', '৫৩', '৫৪', '৫৫', '৫৬', '৫৭', '৫৮', '৫৯', '৬০', '৬১', '৬২', '৬৩',
                        '৬৪', '৬৫', '৬৬', '৬৭', '৬৮', '৬৯', '৭০','৭১','৭২','৭৩','৭৩','৭৪','৭৫','৭৬','৭৭','৭৮','৭৯','৮০','৮১','৮২','৮৩','৮৪','৮৫','৮৬','৮৭','৮৮','৮৯','৯০','৯১','৯২','৯৩','৯৪','৯৫','৯৬','৯৭','৯৮','৯৯','১০০','১০১','১০২','১০৩','১০৪','১০৫','১০৬','১০৭','১০৮','১০৯','১১০','১১১']

chapter_title_ordinal = ['শুন্য', 'প্রথম', 'দ্বিতীয়', 'তৃতীয়', 'চতুর্থ', 'পঞ্চম', 'ষষ্ঠ', 'সপ্তম', 'অষ্টম', 'নবম',
                         'দশম', 'একাদশ', 'দ্বাদশ', 'ত্রয়োদশ', 'চতুর্দশ', 'পঞ্চদশ', 'ষোড়শ', 'সপ্তদশ', 'অষ্টাদশ',
                         'ঊনবিংশ', 'বিংশ', 'একবিংশ', 'দ্বাবিংশ', 'ত্রয়োবিংশ', 'চতুর্বিংশ', 'পঞ্চবিংশ', 'ষট্‌বিংশ',
                         'সপ্তবিংশ', 'অষ্টাবিংশ', 'ঊনত্রিংশ', 'ত্রিংশ', 'একত্রিংশ', 'দ্বাত্রিংশ', 'ত্রয়োত্রিংশ',
                         'চতুর্ত্রিংশ', 'চতুর্ত্রিংশ', 'ষট্‌ত্রিংশ', 'সপ্তত্রিংশ', 'অষ্টাত্রিংশ', 'ঊনচত্বারিংশ',
                         'চত্বারিংশ', 'একচত্বারিংশ', 'দ্বিচত্বারিংশ', 'ত্রয়শ্চত্বারিংশ', 'চতুঃচত্বারিংশ',
                         'পঞ্চচত্বারিংশ', 'ষট্‌চত্বারিংশ', 'সপ্তচত্বারিংশ', 'অষ্টচত্বারিংশ', 'ঊনপঞ্চাশৎ', 'পঞ্চাশৎ', '৫১', '৫২', '৫৩', '৫৪', '৫৫', '৫৬', '৫৭', '৫৮', '৫৯', '৬০', '৬১', '৬২', '৬৩',
                        '৬৪', '৬৫', '৬৬', '৬৭', '৬৮', '৬৯', '৭০','৭১','৭২','৭৩','৭৩','৭৪','৭৫','৭৬','৭৭','৭৮','৭৯','৮০','৮১','৮২','৮৩','৮৪','৮৫','৮৬','৮৭','৮৮','৮৯','৯০','৯১','৯২','৯৩','৯৪','৯৫','৯৬','৯৭','৯৮','৯৯','১০০','১০১','১০২','১০৩','১০৪','১০৫','১০৬','১০৭','১০৮','১০৯','১১০','১১১']

chapter_title_word = ['শুন্য', 'এক', 'দুই', 'তিন', 'চার', 'পাঁচ', 'ছয়', 'সাত', 'আট', 'নয়', 'দশ', 'এগারো', 'বারো',
                      'তেরো', 'চৌদ্দ', 'পনেরো', 'ষোল', 'সতেরো', 'আঠারো', 'উনিশ', 'কুড়ি', 'একুশ', 'বাইশ', 'তেইশ',
                      'চব্বিশ', 'পঁচিশ', 'ছাব্বিশ', 'সাতাশ', 'আঠাশ', 'ঊনত্রিশ', 'ত্রিশ', 'একত্রিশ', 'বত্রিশ', 'তেত্রিশ',
                      'চৌত্রিশ', 'পঁয়ত্রিশ', 'ছত্রিশ', 'সাঁয়ত্রিশ', 'আটত্রিশ', 'ঊনচল্লিশ', 'চল্লিশ', 'একচল্লিশ',
                      'বিয়াল্লিশ', 'তেতাল্লিশ', 'চুয়াল্লিশ', 'পঁয়তাল্লিশ', 'ছেচল্লিশ', 'সাতচল্লিশ', 'আটচল্লিশ',
                      'ঊনপঞ্চাশ', 'পঞ্চাশ', 'একান্ন', 'বাহান্ন', 'তিপ্পান্ন', 'চুয়ান্ন', 'পঞ্চান্ন', 'ছাপ্পান্ন', 'সাতান্ন''আটান্ন', 'ঊনষাট', 'ষাট', 'একষট্টি', 'বাষট্টি', 'তেষট্টি',
                        '৬৪', '৬৫', '৬৬', '৬৭', '৬৮', '৬৯', '৭০','৭১','৭২','৭৩','৭৩','৭৪','৭৫','৭৬','৭৭','৭৮','৭৯','৮০','৮১','৮২','৮৩','৮৪','৮৫','৮৬','৮৭','৮৮','৮৯','৯০','৯১','৯২','৯৩','৯৪','৯৫','৯৬','৯৭','৯৮','৯৯','১০০','১০১','১০২','১০৩','১০৪','১০৫','১০৬','১০৭','১০৮','১০৯','১১০','১১১']


def get_format(chapter_url):
    response = requests.get(chapter_url)
    soup = BeautifulSoup(response.text, features="lxml")
    chapter = soup.find('span', class_="current-item").string

    if chapter[:4] == '০১. ' or chapter[:4] == '০০. ' or chapter[:4] == '১.০১':
        chapter_title_format = 'first'
    elif chapter[:3] == '১. ' or chapter[:3] == '০. ' or chapter[:3] == '১.১':
        chapter_title_format = 'one'
    elif chapter[-4:] == '– ০১' or chapter[-4:] == '– ০০':
        chapter_title_format = 'last'
    else:
        chapter_title_format = 'full'

    title = soup.find('h1', class_='page-header-title').string

    if '(' in title:
        i = title.find('(')
        title = title[:i-1]

    # print("BOOK : " + title)
    # print("format : " + chapter_title_format)

    # if input() == 'n':
        # title = input()

    return chapter_title_format, title


def parse_chapter(chapter_url, chapter_format='full', book_title='book', i=1,serial=1,flag=False):
    response = requests.get(chapter_url)
    soup = BeautifulSoup(response.text, features="lxml")

    title = soup.find('h1', class_='page-header-title').string
    chapter = soup.find('span', class_="current-item").string
    # chapter_number = str(i)
    # chapter_text = chapter

    try:
        previous_url = soup.find('a', rel='prev').get('href')
        # print(previous_url)
    except:
        previous_url = ''
        # print("First Chapter ... No previous Link")

    try:
        next_url = soup.find('a', rel='next').get('href')
        # print(next_url)
    except:
        next_url = ''
        # print("Last Chapter ... No next link")

    if chapter_format == 'first':
        (chapter_number, chapter_text) = (chapter[:2], chapter[4:])
    elif chapter_format == 'one':
        (chapter_number, chapter_text) = (chapter[0], chapter[3:])
    elif chapter_format == 'last' and next_url != '':
        (chapter_number, chapter_text) = (chapter[-2:], chapter[:-4])
    elif chapter_format == 'last' and next_url == '':
        chapter = chapter.replace(' (শেষ)', '')
        (chapter_number, chapter_text) = (chapter[-2:], chapter[:-4])
    else:
        # chapter_number, chapter_text = str(i).zfill(2), chapter
        chapter_number, chapter_text = '', chapter

    try:
        soup.find("div", class_='scriptlesssocialsharing').decompose()
    except:
        pass

    try:
        soup.find("img").decompose()
    except:
        pass

    try:
        soup.find("div", class_='mbfp-btn').decompose()
    except:
        pass

    try:
        soup.find("svg", class_='mbfp-icon').decompose()
    except:
        pass


    body = soup.find("div", class_="entry-content")
    body['class'] = 'body-style'

    # flag = True
    file_name = 'section' + str(i).zfill(4) + ".xhtml"
    # file_name = file_name.replace(' ', '_')

    # ................................................................................

    if i <110:
        ch_num = chapter_title_ordinal[i]
    else:
        ch_num = str(i)
    # ................................................................................
    if chapter_format == "first" or chapter_format == "one":
        temp_head = chapter_text
        temp_number = ch_num
        temp_title = chapter_text

    elif chapter_format == "last" or flag == True:
        temp_head = book_title
        temp_number = book_title
        temp_title = ch_num

    # elif chapter_format == 'full':
    else:
        temp_head = chapter_text
        temp_number = ch_num
        temp_title = chapter_text

    # ................................................................................

    temp =''
    # if temp_number:
    #     temp += r'<h2 class="right sigil_not_in_toc hh1">' + temp_number + '</h2>'
    temp += r'<br/><h1 class="chapter darkblue hh1">' + temp_title + '</h1>\n<hr class="full2"/>\n<br/><br/>'

    temp += str(body).replace("</p>", "</p><br/>")

    global t,c,fn
    fn.append(file_name)
    t.append(temp_title)
    c.append(temp)
    return next_url


def from_first_chapter(first_chapter_url, i=1, book='', writer='UNKNOWN',serial=1,flag = False):
    url = first_chapter_url

    chapter_format, book_title = get_format(url)

    # with open("title.txt",'a',encoding='utf-8') as file:
    #     file.write(book_title+"\n")

    while url != '':
        # print(f'{str(i).zfill(2)} {url}')
        url = parse_chapter(url, chapter_format, book, i,serial,flag)
        i += 1



def parse_book(book_url, i=1, book='', writer='UNKNOWN',serial=1,flag =False):
    response = requests.get(book_url)
    soup = BeautifulSoup(response.text, features="lxml")

    first_chapter_url = soup.find(class_="entry-more-link").get('href')

    from_first_chapter(first_chapter_url, i, book, writer,serial,flag)

def parse_book_from_first_chapter(book_url, i=1, book='', writer='UNKNOWN',serial=1,flag =False):
    from_first_chapter(book_url, i, book, writer,serial,flag)

# parse_book(book_url, i=first_chapter, book=book_name, writer=author_name,serial = 1)
parse_book_from_first_chapter(book_url, i=first_chapter, book=book_name, writer=author_name,serial = 1)


# Generate Cover
#cover
if not path.exists("res.zip"):
    !wget 'https://github.com/AhmedNazir/eBanglaLibrary/raw/main/res.zip'
    
    import zipfile
    with zipfile.ZipFile('res.zip', 'r') as zip_ref:
        zip_ref.extractall()

    from google.colab import output
    output.clear()

print('COMPLETE')

#generate cover
import requests

bbn = requests.get('http://bot123.atwebpages.com/converter.php?unicode=' + book_name).text
bwn = requests.get('http://bot123.atwebpages.com/converter.php?unicode=' + author_name).text

from PIL import Image, ImageFont, ImageDraw 

img = Image.open("res/cover.png")
draw = ImageDraw.Draw(img)

font = ImageFont.truetype("res/KarnaphuliMJBold.ttf",size = 70)
draw.text(xy = (350,50),text = bbn, fill = (0,0,0), font=font)
draw.text(xy = (200,900),text = bwn, fill = (0,0,0), font=font)

img.save('cover.png')



# EBUP Generator

book = epub.EpubBook()
toc=[]

if __name__ == '__main__':
    book = epub.EpubBook()

    #Metadata
    book.set_identifier(str(random.randint(1000000000, 10000000000)))
    book.set_language('bn')
    book.set_title(book_name)
    book.add_author(author_name)
    book.add_metadata('DC', 'publisher', publisher)

    # import css file
    css_text = open('res/style.css','r').read()

    # define css file
    style = epub.EpubItem(uid="style_nav", file_name="style/style.css", media_type="text/css", content=css_text)
    book.add_item(style)


    # add cover image
    book.set_cover("cover.png", open('cover.png', 'rb').read())
    book.spine.append('cover')



    #Title Page
    #embedded image
    
    # load Image file
    img1 = Image.open('res/image.png')  # 'image1.jpeg' should locate in current directory for this example
    b = io.BytesIO()
    img1.save(b, 'png')
    b_image1 = b.getvalue()

    # define Image file path in .epub
    image1_item = epub.EpubItem(uid='image', file_name='images/design.png', media_type='image/png', content=b_image1)

    # add Image file
    book.add_item(image1_item)

    temp = '''   
        <h1 class="hide">বই পরিচিতি</h1>
        <div class="coverbox">
        <h1 class="coverh1">dummy-title</h1>

        <hr class="full2" />
        <br />
        <h2 class="cb">dummy-writer</h2>
        <hr />

        <p class="cb"><img src="images/design.png" width="80" height="81" alt="image of colophon"></img>
            <hr />
            <p class="cb">dummy-publisher</p><br /><br />

        </p>
        </div>
        '''

    temp = temp.replace('dummy-title', book_name)
    temp = temp.replace('dummy-writer', author_name)
    temp = temp.replace('dummy-publisher', publisher)

    c1 = epub.EpubHtml(title='বই পরিচিতি', file_name='name.xhtml', lang='en')
    c1.content = temp
    c1.add_item(style)
    book.spine.append(c1)
    book.add_item(c1)
    toc.append(epub.Link('name.xhtml', 'বই পরিচিতি', 'title' ))


    

    #Index Page
    book.spine.append('nav')


    #reader Page
    temp = ''' 
        <h1 class="cover-writer-title center">Recommended<br/>EPUB Reader</h1>

        <br /><br />
        <p class="red"><b>Android</b></p>
        <p class="green"><b>#  Google Play Books</b></p>
        <p class="green"><b># Lithium Epub Reader</b></p>

        <br /><br />
        <p class="red"><b>Windows PC</b></p>
        <p class="green"><b># Calibre</b></p>

        <br /><br />
        <p class="red"><b>Apple</b></p>
        <p class="green"><b># iBooks</b></p>
    '''

    c1 = epub.EpubHtml(title='EPUB Reader', file_name='reader.xhtml', lang='en')
    c1.content = temp
    c1.add_item(style)
    book.spine.append(c1)
    book.add_item(c1)
    toc.append(epub.Link('reader.xhtml', 'EPUB Reader', 'reader' ))




    #chapter Page

    for i in range(len(fn)):
        temp = c[i]
        c1 = epub.EpubHtml(title=book_name, file_name=fn[i], lang='en')
        c1.content = temp
        c1.add_item(style)
        book.add_item(c1)
        book.spine.append(c1)
        toc.append(epub.Link(fn[i], t[i], fn[i].split('.')[0]))




    #Finish Page
    temp = ''' 
        <h1 class="hh1 finish red">সমাপ্ত </h1>
        <br />
        <p class="finish green">বইটি ভাল লাগলে অবশ্যই কিনবেন</p>
        <p class="finish green">অথবা</p>
        <p class="finish green">অন্যকে উপহার দিবেন</p>
    '''

    c2 = epub.EpubHtml(title=u'সমাপ্ত', file_name='section9999.xhtml', lang='en')
    c2.content = temp
    c2.add_item(style)
    book.add_item(c2)
    book.spine.append(c2)
    toc.append(epub.Link('section9999.xhtml', u'সমাপ্ত', 'Finish'))


    book.toc = tuple(toc)

    # add navigation files
    book.add_item(epub.EpubNcx())
    book.add_item(epub.EpubNav())

    # create epub file
    epub.write_epub(book_name+' - ' + author_name +'.epub', book, {})


```
